```python
import pandas as pd
import csv
```

```python
trend_csv = "trend_list_15.csv"
trend_df = pd.read_csv(trend_csv)
trend_df = trend_df.drop(trend_df.columns[0], axis=1)
trend_df
```

```python
trend_list = []
for row in trend_df.iterrows():
    trend_list.append(row[1][0])
    trend_list.append(row[1][1])
    trend_list.append(row[1][2])
    trend_list.append(row[1][3])
    trend_list.append(row[1][4])
    trend_list.append(row[1][5])
    trend_list.append(row[1][6])
    trend_list.append(row[1][7])
    trend_list.append(row[1][8])
    trend_list.append(row[1][9])
    trend_list.append(row[1][10])
    trend_list.append(row[1][11])
    trend_list.append(row[1][12])
    trend_list.append(row[1][13])
    trend_list.append(row[1][14])
# trend_list
```

```python
trend_list_df = pd.DataFrame(trend_list)

trend_list_df = trend_list_df.rename(columns = {0: "Trends"}, inplace = False)
trend_list_df
```

```python
# trend_list_df.to_csv("trend_list2.csv", index = False, header = True)
```

```python
unique_trends = trend_list_df['Trends'].value_counts().rename_axis('unique_values').reset_index(name='counts')
unique_trends.to_csv("unique_trends.csv", index = False, header = True)
unique_trends
```
